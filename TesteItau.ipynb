```python
import pandas as pd
import matplotlib.pyplot as plt
arquivo = '/content/drive/MyDrive/DadosTratadosItau.csv'
df = pd.read_csv(arquivo)
df
```

```python
# Analisando a base de dados, qual o tipo de público (considerando gênero e idade)

def idade_sexo(linha):
  if linha["Customer Age "] >= 0 and linha["Customer Age "] <= 19 and linha["Gender"] == "Female":
    return "Mulher entre 0 e 19 anos"
  elif linha["Customer Age "] >= 20 and linha["Customer Age "] <= 35 and linha["Gender"] == "Female":
    return "Mulher entre 20 e 35 anos"
  elif linha["Customer Age "] >= 36 and linha["Customer Age "] <= 50 and linha["Gender"] == "Female":
    return "Mulher entre 36 e 50 anos"
  elif linha["Customer Age "] >=50 and linha["Gender"] == "Female":
    return "Mulher com 50 anos ou mais"
  if linha["Customer Age "] >= 0 and linha["Customer Age "] <= 19 and linha["Gender"] == "Male":
    return "Homem entre 0 e 19 anos"
  elif linha["Customer Age "] >= 20 and linha["Customer Age "] <= 35 and linha["Gender"] == "Male":
    return "Homem entre 20 e 35 anos"
  elif linha["Customer Age "] >= 36 and linha["Customer Age "] <= 50 and linha["Gender"] == "Male":
    return "Homem entre 36 e 50 anos"
  elif linha["Customer Age "] >=50 and linha["Gender"] == "Male":
    return "Homem com 50 anos ou mais"

print('Observando o gráfico e os valores apresentados, é possivel perceber que mulheres e homens acima de 50 anos possuem mais tendência a compra.')

df["idade_genero"] = df.apply(lambda linha: idade_sexo(linha), axis=1)
df["idade_genero"].value_counts().plot(kind="bar", title='Gráfico Clientes Gênero e Idade')
df["idade_genero"].value_counts()
```

```python
# e o canal ideal para vender determinado produto?
# ELETRONICO

def eletronico(linha):
  if linha["Product Category"] == 'Electronics' and linha["Source"] == 'Instagram Campign':
    return 'Produtos eletrônicos Instagram'
  elif linha["Product Category"] == 'Electronics' and linha["Source"] == 'FaceBook campaign':
    return 'Produtos eletrônicos Facebook'
  elif linha["Product Category"] == 'Electronics' and linha["Source"] == 'Organic Search':
    return 'Produtos eletrônicos Orgânico'
  elif linha["Product Category"] == 'Electronics' and linha["Source"] == 'SEM':
    return 'Produtos eletrônicos Sem nenhuma'

print("Observando os valores e o gráfico abaixo, é possivel analisar que o canal ideal para vender produtos eletrônicos é o Instagram")
df["Canal Eletronico"] = df.apply(lambda linha: eletronico(linha), axis=1)
df["Canal Eletronico"].value_counts().plot(kind="bar", title='Gráfico Canal de Produtos Eletronicos')
df["Canal Eletronico"].value_counts()
```

```python
# e o canal ideal para vender determinado produto?
# HOME

def home(linha):
  if linha["Product Category"] == 'Home' and linha["Source"] == 'Instagram Campign':
    return 'Produtos para Casa Instagram'
  elif linha["Product Category"] == 'Home' and linha["Source"] == 'FaceBook campaign':
    return 'Produtos para Casa Facebook'
  elif linha["Product Category"] == 'Home' and linha["Source"] == 'Organic Search':
    return 'Produtos para Casa Orgânico'
  elif linha["Product Category"] == 'Home' and linha["Source"] == 'SEM':
    return 'Produtos para Casa Sem nenhuma'

print("Observando os valores e o gráfico abaixo, é possivel analisar que o canal ideal para vender produtos para casa(home) é o Instagram")
df["Canal Home"] = df.apply(lambda linha: home(linha), axis=1)
df["Canal Home"].value_counts().plot(kind="bar", title='Gráfico Canal de Produtos para Casa(Home)')
df["Canal Home"].value_counts()
```

```python
# e o canal ideal para vender determinado produto?
# Clothing

def clothing(linha):
  if linha["Product Category"] == 'Clothing' and linha["Source"] == 'Instagram Campign':
    return 'Produtos de Vestuário Instagram'
  elif linha["Product Category"] == 'Clothing' and linha["Source"] == 'FaceBook campaign':
    return 'Produtos de Vestuário Facebook'
  elif linha["Product Category"] == 'Clothing' and linha["Source"] == 'Organic Search':
    return 'Produtos de Vestuário Orgânico'
  elif linha["Product Category"] == 'Clothing' and linha["Source"] == 'SEM':
    return 'Produtos de Vestuário Sem nenhuma'

print("Observando os valores e o gráfico abaixo, é possivel analisar que o canal ideal para vender produtos para Vestuário(Clothing) é o Instagram")
df["Canal Clothing"] = df.apply(lambda linha: clothing(linha), axis=1)
df["Canal Clothing"].value_counts().plot(kind="bar", title='Gráfico Canal de Produtos de Vestuário(Clothing)')
df["Canal Clothing"].value_counts()
```

```python
# e o canal ideal para vender determinado produto?
# Books

def books(linha):
  if linha["Product Category"] == 'Books' and linha["Source"] == 'Instagram Campign':
    return 'Produtos Livros Instagram'
  elif linha["Product Category"] == 'Books' and linha["Source"] == 'FaceBook campaign':
    return 'Produtos Livros Facebook'
  elif linha["Product Category"] == 'Books' and linha["Source"] == 'Organic Search':
    return 'Produtos Livros Orgânico'
  elif linha["Product Category"] == 'Books' and linha["Source"] == 'SEM':
    return 'Produtos Livros Sem nenhuma'

print("Observando os valores e o gráfico abaixo, é possivel analisar que o canal ideal para vender produtos para Livros(Books) é o Instagram")
df["Canal Books"] = df.apply(lambda linha: books(linha), axis=1)
df["Canal Books"].value_counts().plot(kind="bar", title='Gráfico Canal de Produtos para Casa(Home)')
df["Canal Books"].value_counts()
```

```python
# 1 - Qual o produto mais vendido considerando os 3 ultimos anos
df['data'] = df['Purchase Date'].apply(pd.to_datetime)
df['Ano'] = df['data'].dt.year

def ano(linha):
  if linha["Ano"] >= 2021 and linha["Product Category"] == "Electronics":
    return 'Produto eletrônico comprado em 2021 ou mais'
  if linha["Ano"] >= 2021 and linha["Product Category"] == "Home":
    return 'Produto de casa comprado em 2021 ou mais'
  if linha["Ano"] >= 2021 and linha["Product Category"] == "Clothing":
    return 'Produto vestuário comprado em 2021 ou mais'
  if linha["Ano"] >= 2021 and linha["Product Category"] == "Books":
    return 'Produto livro comprado em 2021 ou mais'

print("Analisando os valores e o gráfico, é possivel verificar que o produto mais vendido nos últimos 3 anos é de Vestuário(Clothing)")
df["Compra >= 2021"] = df.apply(lambda linha: ano(linha), axis=1)
df["Compra >= 2021"].value_counts().plot(kind="bar", title='Gráfico Compra de Produtos >= 2021')
df["Compra >= 2021"].value_counts()
```

```python
# 2.1 - Qual o produto mais caro?
print('Todos os produtos(categoria) possuem o mesmo valor de produto mais caro')
df.groupby("Product Category")["Product Price"].max()
# obs: o exerc pede o produto em si, mas a database disponibilizada possuí apenas a tabela "Product Category"
```

```python
# 2.2 - E o mais barato?
print('Todos os produtos(categoria) possuem o mesmo valor de produto mais barato')
df.groupby("Product Category")["Product Price"].min()
```

```python
# 3 - Qual a categoria de produto mais vendida? E a menos vendida?
df.groupby("Product Category")["Product Price"].sum().plot(kind='bar', title = 'Soma de valores das categorias de produtos')

df1 = df.groupby("Product Category")["Product Price"].sum()
df1

# De acordo com o gráfico abaixo e os valores apresentados na soma da categoria, é possível verificar que o produto(categoria) mais vendido é o "Clothing"
# E o produto(categoria) menos vendido é o "Home"
```

```python
# 4 - Qual o produto com o melhor e o pior NPS?
df.groupby("Product Category")["NPS"].mean()
# Observando os valores abaixo, a média de NPS por categoria de produto que possuí o maior NPS é o "Home", enquanto o que possuí o menor NPS é o "Electronics"
```
